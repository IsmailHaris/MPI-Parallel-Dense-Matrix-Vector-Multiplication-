Parallel MPI vs Sequential Matrix-Vector Multiplication

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

We fix the size of the matrix to 10 000 x 10 000 and compare the Speed Up and the Computation time for different numbers of processors. I have at time of writing 4 procs at disposal.
The Comparision is made between the Sequential version and either the results of the worst processor or the average processor performance.
Note that I only compare time performances since the result error is always zero. 

```
#nx fixed at 100 so nrows =10000

ncores =[1,2,3,4,6,8]
time_ncore_mpi_max = [0.0333,0.0292, 0.0305, 0.0307, 0.0308, 0.033]
time_ncore_mpi_average = [0.0333,0.0292, 0.028, 0.026, 0.029, 0.0267]
time_ncore_sequential = [0.04509,0.049, 0.0578, 0.0585, 0.064, 0.059]
error_ncores = [0, 0,0, 0, 0, 0]
Speedup_max = [time_ncore_sequential[i]/time_ncore_mpi_max[i] for i in range(len(time_ncore_mpi_max))]
Speedup_ave = [time_ncore_sequential[i]/time_ncore_mpi_average[i] for i in range(len(time_ncore_mpi_average))]
```

```
plt.plot(ncores,time_ncore_mpi_max, label ="MPI")
plt.plot(ncores, time_ncore_sequential, label ='Sequential')
#plt.plot(ncores, erreur_ncores)
plt.legend()
plt.title("Comparison Time Parallel MPI Computation (max core) vs Time Sequential for a matrix 10 000 x 10 000")
plt.xlabel("Ncores")
plt.ylabel("Time (s)")
plt.show()
```

```
plt.plot(ncores,time_ncore_mpi_average, label ="MPI")
plt.plot(ncores, time_ncore_sequential, label ='Sequential')
#plt.plot(ncores, erreur_ncores)
plt.legend()
plt.title("Comparison Time Parallel MPI Computation (average core) vs Time Sequential for a matrix 10 000 x 10 000")
plt.xlabel("Ncores")
plt.ylabel("Time (s)")
plt.show()
```

```
plt.plot(ncores,Speedup_ave, label ="MPI")
plt.plot(ncores, [1 for i in ncores], label ='Sequential')
#plt.plot(ncores, erreur_ncores)
plt.legend()
plt.title("SpeedUp MPI (average core) vs Sequential  for a 10 000 x 10 000 matrix ")
plt.xlabel("Ncores")
plt.ylabel("Speedup ")
plt.show()
```

```
plt.plot(ncores,Speedup_max, label ="MPI")
plt.plot(ncores, [1 for i in ncores], label ='Sequential')
#plt.plot(ncores, erreur_ncores)
plt.legend()
plt.title("SpeedUp MPI (max core) vs Sequential  for a 10 000 x 10 000 matrix ")
plt.xlabel("Ncores")
plt.ylabel("Speedup ")
plt.show()
```
